## Find total spills per year

I start by using the ny_spills dataframe to create a new one just including the past ten years. I'm starting from ny_spills because I want to access all of the columns again, which I got rid of when I made a new dataframe after that 

```python
entire_df_last_ten = ny_spills[ny_spills['Spill Date'] >= '2013-01-01'] 
entire_df_last_ten
```

### Once again, I'll get rid of everything that's not a county in New York City

```python
new_york_city_last_ten = entire_df_last_ten[(entire_df_last_ten['County'] == 'Bronx') | (entire_df_last_ten['County'] == 'Kings') | (entire_df_last_ten['County']== 'New York') | (entire_df_last_ten['County']=='Queens')| (entire_df_last_ten['County']=='Richmond')]
new_york_city_last_ten
```

### Group spills by the Spill Date

```python
annual_spills = entire_df_last_ten.groupby([pd.Grouper(key='Spill Date', axis=0, freq='A')]).count()
```

```python
annual_spills
```

```python
print(annual_spills.columns)
```

I can get rid of every column but Spill Date and Spill Number

```python
new_annual_spills = annual_spills[['Spill Number']].reset_index()
new_annual_spills
```

### Rename columns

```python
new_annual_spills.rename(columns={'Spill Number': 'spills'}, inplace=True)
new_annual_spills
```

I want to create a new column with just the year

```python
new_annual_spills['year'] = new_annual_spills['Spill Date'].dt.year
new_annual_spills
```

```python
new_annual_spills = new_annual_spills[['year', 'spills']].copy()
new_annual_spills
```

Lastly, I can make a chart that visualizes the amount of spills in each of the past 10 years

```python
alt.Chart(new_annual_spills).mark_bar().encode(
    x='year:O',
    y='spills'
).properties(
    title='New York City toxic spills by year'
)
```
