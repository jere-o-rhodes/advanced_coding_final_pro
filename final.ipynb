### Convert spill date, close date and received date to date time dtype

```python
ny_spills['Spill Date'] = pd.to_datetime(ny_spills['Spill Date'])
ny_spills['Received Date'] = pd.to_datetime(ny_spills['Received Date'])
ny_spills['Close Date'] = pd.to_datetime(ny_spills['Close Date'])
```

```python
ny_spills['Spill Date']
```

### Create a new column that shows how much time it took for a case to be closed

```python
ny_spills['Time_to_close'] = ny_spills['Close Date'] - ny_spills['Received Date']
```

```python
ny_spills
```

```python
ny_spills['Time_to_close'].mean()
```

```python
ny_spills['Time_to_close'].median()
```

```python
ny_subset = ny_spills[['Spill Number', 'County', 'Spill Date', 'Close Date', 'Quantity', 'Units', 'Time_to_close', 'Material Name']]
```

```python
ny_subset
```

### Only want spills from New York City, which has five counties: Bronx, Kings, Queens, Richmond and New York

```python
just_nyc = ny_subset[(ny_subset['County'] == 'Bronx') | (ny_subset['County'] == 'Kings') | (ny_subset['County'] == 'New York') | (ny_subset['County'] == 'Queens') | (ny_subset['County'] == 'Richmond')]
just_nyc
```

```python
just_nyc.sort_values(by=['Time_to_close'], ascending=False).head(10)
```

### Only want spills from the past 10 years

```python
last_ten_years = just_nyc[just_nyc['Spill Date'] >= '2013-01-01'] 
last_ten_years
```

```python
last_ten_years.sort_values(by=['Time_to_close'], ascending=False).head(10)
```

```python
last_ten_years.sort_values(by=['Quantity'], ascending=False).head(10)
```
